```python
import gsidiag as gd
import hvplot.pandas
import panel as pn

pn.extension()
```

```python
file = gd.read_diag('../data/diag_conv_01.2020031112')
```

```python
_kx_values = file.overview()

var = pn.widgets.Select(name='Variable', value=file.varNames[0], options=file.varNames)
kx = pn.widgets.Select(name='kx', value=_kx_values[var.value][0], options=_kx_values[var.value])

all_kx_by_level = pn.widgets.Toggle(name='by Level', value=False, button_type='success')
kx_by_level = pn.widgets.Toggle(name='by Level', value=False, button_type='success')

@pn.depends(var.param.value, watch=True)
def _update_kx_values(select_var):
    kx_values = _kx_values[select_var]
    kx.options = kx_values
    kx.value = kx_values[0]

@pn.depends(var, kx, all_kx_by_level, kx_by_level)
def plotPcount(var, kx, all_kx_by_level, kx_by_level):
    return gd.plot_diag.pcount_hvplot(file, var, kx, all_kx_by_level, kx_by_level)

pn.Row(
    pn.Row(
        pn.Card(pn.Column(var, all_kx_by_level), title='By Variable and kx'), 
        pn.Card(pn.Column(kx_by_level, kx), title='By Level and kx')
    ), 
    plotPcount
).servable()
```

```python
_kx_values = file.overview()

var = pn.widgets.Select(name='Variable', value=file.varNames[0], options=file.varNames)
lev = pn.widgets.Select(name='Level', value=file.zlevs[0], options=file.zlevs)
kx = pn.widgets.Select(name='kx', value=_kx_values[var.value][0], options=_kx_values[var.value])
iuse = pn.widgets.Select(name='iuse', value=1, options=[-1, 1])

@pn.depends(var.param.value, watch=True)
def _update_kx_values(select_var):
    kx_values = _kx_values[select_var]
    kx.options = kx_values
    kx.value = kx_values[0]

@pn.depends(var, kx, lev, iuse)
def plotPtmap(var, kx, lev, iuse):
    return gd.plot_diag.ptmap_hvplot(file, var, kx, lev, iuse)

pn.Row(
    pn.Column(var, lev, iuse, kx), plotPtmap
).servable()
```

```python
_kx_valuesm = file.overview()

varm = pn.widgets.Select(name='Variable', value=file.varNames[0], options=file.varNames)
levm = pn.widgets.Select(name='Level', value=file.zlevs[0], options=file.zlevs)
kxm = pn.widgets.MultiChoice(name='kx', value=_kx_valuesm[varm.value], options=_kx_valuesm[varm.value], solid=False)
iusem = pn.widgets.Select(name='iuse', value=1, options=[-1, 1])

@pn.depends(varm.param.value, watch=True)
def _update_kx_values(select_varm):
    kx_valuesm = _kx_valuesm[select_varm]
    kxm.options = kx_valuesm
    kxm.value = kx_valuesm

@pn.depends(varm, kxm, levm, iusem)
def plotPtmapM(varm, kxm, levm, iusem):
    return gd.plot_diag.ptmap_hvplot_multi(file, varm, kxm, levm, iusem)

pn.Row(
    pn.Column(varm, levm, iusem, kxm, height=300), plotPtmapM
).servable()
```
